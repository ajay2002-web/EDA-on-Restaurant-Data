### Replacing 0's with NaN

```python
df[['Longitude', 'Latitude','Average Cost for two']]=df[['Longitude', 'Latitude','Average Cost for two']].replace(0,np.NaN)
```

```python
missing = msno.matrix(df)
missing.set_title("Missing data for Restaurant Dataset",fontsize=20)
```

### Visualising the number of restaurants using Pie Chart  
We can see below that almost 70% of the restaurants are located in Delhi. This information can be useful in further analysis of the data.

```python
# Enlisting the top 5 cities from the data
plt.figure(figsize=(10,10))
city_names=df.City.value_counts().index
city_values=df.City.value_counts().values
plt.pie(city_values[:5],labels=city_names[:5],autopct='%1.2f%%');
```

### Aggregating Data by their Rating

```python
ratings=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

### From the above observation :  
* When rating is between 4.5 to 4.9 it means Excelent  
* When rating is between 4.0 to 4.4 it means Very Good  
* When ratingf is between 3.5 to 3.9 it means Good  
* When ratingf is between 2.5 to 3.5 it means Average  
* When ratingf is between 1.5 to 2.5 it means Poor   
* When rating is 0 it means it is not rated  
We can also see that the number of reastaurants with 0 rating is the highest. So, it is a good practice to further breakdown this category

```python
plt.figure(figsize=(20,10))
sns.barplot(x="Aggregate rating", y="Rating Count",data=ratings);
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="Aggregate rating", y="Rating Count",data=ratings,hue="Rating color");
```
